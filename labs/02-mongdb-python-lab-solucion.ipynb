## Exercicio 3. Consulta con filtro
Obtén os códigos postais cunha poboación superior a 30.000 habitantes.
Amosa o código postal, a cidade e a poboación.

```python
for doc in coleccion.find({"pop":{"$gte":30000}},{"zip":1,"city":1,"pop":1,"_id":0}):
    print(doc)
```

## Exercicio 5. Ordenación e límite
Obtén os 10 códigos postais máis poboados da colección, ordenados de
maior a menor poboación.

```python
for doc in coleccion.find({}).sort({"pop":-1}).limit(10):
    print(doc)
```

## Exercicio 6. Actualización de documentos
Incrementa en 500 habitantes a poboación de todos os códigos postais
do estado "TX". Indica cantos documentos foron modificados.

```python
res = coleccion.update_many({"state":"TX"}, {"$inc": {"pop":-500}})
print ("Modificados: ",res.modified_count)
```

## Exercicio 7. Eliminación de documentos
Elimina todos os documentos cunha poboación inferior a 100 habitantes
e indica cantos documentos foron eliminados.

```python
res = coleccion.delete_many({"pop":{"$lt":100}})
print ("Elminados: ", res.deleted_count)
```

## Exercicio 8. Aggregation Pipeline
Usando un pipeline de agregación, obtén para cada estado:
- a poboación total
- o número de códigos postais
Ordena o resultado por poboación total descendente.

```python
pipeline = [
    {"$group": {
        "_id":"$state",
        "poboacion_total":{"$sum":"$pop"},
        "num_zips": {"$sum":1}
    }},
    {"$sort": {"poboacion_total":-1}},
    {"$project":{
        "_id":0,
        "estado":"$_id",
        "poboacion_total":1,
        "num_zips":1
        
    }}
]
cursor = coleccion.aggregate(pipeline)
```

## Exercicio 9. Conversión a pandas
Converte o resultado da agregación do exercicio anterior nun DataFrame
de pandas. Renomea o campo _id a "estado" e amosa as primeiras filas.

```python
import pandas as pd

df = pd.DataFrame(list(cursor))
df.head()
```

## Exercicio 10. Análise e visualización
Representa nunha gráfica de barras os 10 estados con maior poboación total.

```python
import matplotlib.pyplot as plt
top10 = df.head(10)

plt.bar(top10["estado"], top10["poboacion_total"])
plt.xticks(rotation=90)
plt.show()
```
